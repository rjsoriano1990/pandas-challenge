```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.dropna()
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
players = purchase_data.loc[:, ["Gender", "SN", "Age"]]

players = players.drop_duplicates()

playercount = players.count()[0]
purchase_data['Player Count'] = playercount
player_count_df = purchase_data[['Player Count']]
player_count_df.head(1)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data['Number of Unique Items'] = purchase_data['Item Name'].nunique()
unique_items = purchase_data[['Number of Unique Items']]

purchase_data['Average Price'] = purchase_data['Price'].mean()
average_price = purchase_data[['Average Price']]


purchase_data['Number of Purchases'] = purchase_data['Purchase ID'].count()
number_of_purchases = purchase_data[['Number of Purchases']]

purchase_data['Total Revenue'] = purchase_data['Price'].sum()
total_revenue = purchase_data[['Total Revenue']]

pd.options.display.float_format = "${:,.2f}".format
total_analysis_df = purchase_data[['Number of Unique Items', 'Average Price', 
                                   'Number of Purchases', 'Total Revenue']]
total_analysis_df.head(1)


```

## Gender Demographics

* Percentage and Count of Female Players

```python
gender_count = players["Gender"].value_counts()
gender_percent = gender_count / playercount * 100
gender_demo_df = pd.DataFrame({"Percentage of Players": gender_percent,
                                 "Total Count": gender_count})

gender_demo_df= gender_demo_df.round(2)
gender_demo_df.head()

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ages_bins = [0,10,15,20,25,30,35,40,999]
age_ranges = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

unique_players = purchase_data.loc[:,["SN","Age"]]
unique_players["Age Ranges"] = pd.cut(unique_players["Age"], ages_bins, labels=age_ranges)
unique_players

age_demo_totals = unique_players["Age Ranges"].value_counts()
age_demo_percents = (age_demo_totals / playercount * 100).round(2)
age_demo = pd.DataFrame({"Total": age_demo_totals,
                        "Percentage of Players": age_demo_percents,
                         })

age_demo.sort_index()
```
